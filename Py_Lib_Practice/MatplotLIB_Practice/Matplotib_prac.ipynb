**TO cover a whole part**

plt.subplot2grid(shape, loc, rowspan=1, colspan=1)

| Argument  | Meaning                                | Example                                   |
| --------- | -------------------------------------- | ----------------------------------------- |
| `shape`   | Size of the overall grid (rows, cols)  | `(4, 4)` → 4 rows, 4 columns              |
| `loc`     | Starting position (row, col)           | `(3, 0)` means 4th row, 1st column        |
| `rowspan` | How many rows the subplot should cover | `2` means span 2 rows vertically          |
| `colspan` | How many columns it should cover       | `4` means span all 4 columns horizontally |

**TEXTS DRAWING**

plt.text(x, y, s, fontdict=None, **kwargs)

| Argument     | Description                                                                  |
| ------------ | ---------------------------------------------------------------------------- |
| **x**        | x-coordinate of the text position                                            |
| **y**        | y-coordinate of the text position                                            |
| **s**        | The string (text) you want to display                                        |
| **fontdict** | (Optional) A dictionary to control font properties (e.g. size, color, style) |
| **kwargs**   | Additional keyword args like alignment, color, rotation, etc.                |



'ha' is short for “horizontal alignment” in Matplotlib.

```python
x = np.linspace(-2,2,50)
y=x**2
plt.plot(x,y)
plt.text(x=0,y=3,fontdict={'family': 'serif', 'size': 12},s='y=x^2',color='orange',ha='center')

plt.xlabel="X LABEL"
plt.ylabel="Y LABEL"

plt.title("PARABOLLA BITCHHH!!!")
plt.grid(True)

plt.show()
```

**ANNOTATE**

plt.annotate(
    s, xy, xytext=None, xycoords='data', textcoords='data',
    arrowprops=None, fontsize=None, color=None, ha='left', va='bottom', **kwargs
)

| Argument       | Meaning                                                                          |
| -------------- | -------------------------------------------------------------------------------- |
| **s**          | The text string to display (like `plt.text`)                                     |
| **xy**         | Tuple `(x, y)` of the **point to annotate**                                      |
| **xytext**     | Tuple `(x, y)` of **where to place the text** (if different from `xy`)           |
| **xycoords**   | Coordinate system for `xy` (`'data'`, `'axes fraction'`, `'figure fraction'`)    |
| **textcoords** | Coordinate system for text (`'data'` by default)                                 |
| **arrowprops** | Dictionary defining arrow style; e.g. `dict(facecolor='black', arrowstyle='->')` |
| **fontsize**   | Font size of the text                                                            |
| **color**      | Text color                                                                       |
| **ha**         | Horizontal alignment of text (`'left'`, `'center'`, `'right'`)                   |
| **va**         | Vertical alignment of text (`'top'`, `'center'`, `'bottom'`)                     |


```python
x=np.linspace(-2,2,50)
y=x**2
plt.plot(x,y,color='red')
plt.annotate("Here's a POINT",(1,1),(0.5,3),arrowprops=dict(facecolor='black',arrowstyle='->'),weight='heavy',color='blue')
plt.show()
```

**BOUNDING BOX**

plt.text(x, y, "Some text", bbox=dict(facecolor='color', edgecolor='color', alpha=..., boxstyle=...))


| Key         | Meaning                                | Example                                       |
| ----------- | -------------------------------------- | --------------------------------------------- |
| `facecolor` | Background color of the box            | `'yellow'`                                    |
| `edgecolor` | Color of the box border                | `'red'`                                       |
| `alpha`     | Transparency (0=transparent, 1=opaque) | `0.3`                                         |
| `boxstyle`  | Shape of the box                       | `'round'`, `'square'`, `'round4'`, `'circle'` |






**ALSO:**



| Key                       | Meaning                                | Example / Explanation                                                                                                                          |
| ------------------------- | -------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------- |
| `boxstyle`                | Shape and style of the box             | `"rarrow,pad=0.3"` → a **right-pointing arrow** with padding 0.3 around the text. Other options: `'round'`, `'round4'`, `'circle'`, `'square'` |
| `pad` (inside `boxstyle`) | Extra space between text and box edges | `0.3` is moderate padding                                                                                                                      |
| `ec`                      | Edge color (border color)              | `"b"` → blue border                                                                                                                            |
| `lw`                      | Line width of the box border           | `2` → medium thickness                                                                                                                         |
| `fc`                      | Face color (background color)          | `"lightblue"`                                                                                                                                  |


```python
x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2
plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "Beautiful point", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black",
                  fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='black',
         ha="center", bbox=bbox_props)

plt.show()
```

**RANDOM SEED GENERATOR**

np.random.seed(seed_value)

1) np.random.seed() sets the “starting point” for the random number generator.

2) This ensures that the sequence of random numbers is reproducible.

3) Without setting a seed, every time you run your code, you’ll get different random numbers.

4) With a fixed seed (like 42), you’ll always get the same random numbers every time.

```python
np.random.seed(42)  # to make this notebook's output reproducible

x, y = np.random.rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 200)
line, = ax.plot(x, np.sin(x))

def update(frame):
    line.set_ydata(np.sin(x + frame/10))  # update y-data
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50)
plt.show()
```

**HISTOGRAMS**

plt.hist(x, bins=10, range=None, density=False, color=None, alpha=1.0, label=None, histtype='bar', edgecolor=None)

| Argument    | Meaning                                                    | Example                           |
| ----------- | ---------------------------------------------------------- | --------------------------------- |
| `x`         | The data you want to plot                                  | `np.random.randn(1000)`           |
| `bins`      | Number of bins or the bin edges                            | `10`, `[0,1,2,3]`                 |
| `range`     | Tuple `(min, max)` to limit the histogram                  | `(0, 5)`                          |
| `density`   | `True` → normalize histogram to form a probability density | `True/False`                      |
| `color`     | Color of bars                                              | `'blue'`                          |
| `alpha`     | Transparency of bars (0–1)                                 | `0.5`                             |
| `label`     | Label for legend                                           | `'My data'`                       |
| `histtype`  | Type of histogram                                          | `'bar'`, `'step'`, `'stepfilled'` |
| `edgecolor` | Color of bar edges                                         | `'black'`                         |


```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
data = np.random.randn(1000)  # 1000 random numbers from normal distribution

# Create histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Histogram Example")

plt.show()
```

# **HEXAGONAL Binning:**

When you have many data points, a normal scatter plot can get messy — too many points overlap.
Hexagonal binning replaces points with hexagonal tiles, where:

Each hexagon’s color or shade represents how many points fall into that bin.

It’s similar to a 2D histogram but with hexagonal cells (better spacing).

<br>

<br>

| Argument            | Description                                                | Example                                          |
| ------------------- | ---------------------------------------------------------- | ------------------------------------------------ |
| `x`, `y`            | The data arrays                                            | `df['Population'], df['Murder Rate']`            |
| `gridsize`          | Number of hexagons in the x-direction (affects resolution) | `gridsize=10` or `gridsize=30`                   |
| `cmap`              | Colormap                                                   | `'Blues'`, `'plasma'`, `'coolwarm'`, `'viridis'` |
| `extent`            | Defines (xmin, xmax, ymin, ymax)                           | `extent=(0, 40000000, 0, 10)`                    |
| `mincnt`            | Minimum count to show a bin                                | `mincnt=1`                                       |
| `reduce_C_function` | Custom function for aggregating within bins                | `reduce_C_function=np.mean` (instead of counts)  |


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random

df = pd.DataFrame(np.random.randn(200,3),columns=['A','B','C'])
df
```

```python
df.plot(kind='hexbin',x='A',y='B',gridsize=15)
```

```python
df.plot(kind='hexbin',x='A',y='B',gridsize=15,cmap="inferno")
```
