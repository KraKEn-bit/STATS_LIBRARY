```python
def add_numb(x,y,z):
  return x+y*z

def func_x(Cols,rows,blocks):
  return np.fromfunction(add_numb,(blocks,rows,Cols))

def func_y(Cols,rows,blocks):
  return np.fromfunction(add_numb,(blocks,rows,Cols))

x=func_x(4,3,2)
y=func_y(3,2,4)

plt.scatter(x,y,label="Random",color="red")
plt.legend()
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.title("SCATTER PLOT")
plt.show()
```

**MATRIX OPERATION- REF/RREF**

```python
import numpy as np
import sympy as sp

A = np.array([[2, 1, -1, 8],
              [-3, -1, 2, -11],
              [-2, 1, 2, -3]])

M = sp.Matrix(A)

sp.pprint(M.echelon_form())
```

```python
sp.pprint(M.rref())
```

```python
M.rank()
```

```python
new_arr=np.array([[2, 1, -1, 8],
                  [-3, -1, 2, -11],
                  [-2, 1, 2, -3]])

row_ech_form = sp.Matrix(new_arr).echelon_form()
print("Row Echelon Form:", row_ech_form)

red_row_ech_form = row_ech_form.rref()
print("Reduced Row Echelon Form:",red_row_ech_form)
```

**MATRIX INVERSE AND PSEUDO INVERSE**

```python
import numpy.linalg as linearalgebra

arr=np.array([[2, 1, -1],
                  [-3, -1, 2],
                  [-2, 1, 2]])

linearalgebra.inv(arr)
```

**DOING PSEUDO INVERSE (Not square matrix)**

```python
arr2= np.array([[2, 1, -1, 8],
                  [-3, -1, 2, -11],
                  [-2, 1, 2, -3]])

linearalgebra.pinv(arr2)
```

**QR DECOMPOSITION/Matrix Decomposition**

```python
import numpy.linalg as linearalgebra
import numpy as np

arr=np.array([[2, 1, -1],
                  [-3, -1, 2],
                  [-2, 1, 2]])

q,r = linearalgebra.qr(arr)
q
```

```python
r
```

```python
linearalgebra.det(arr)
```

**EIGENVALUES and EIGENVECTORS**

```python
EigenValues,EigenVectors = linearalgebra.eig(arr)
print("EigenValues: ",EigenValues)
print("Eigenvectors: ",EigenVectors)
```

**SVD (Singular Value Decomposition)**

```python
arr=np.array([[2, 1, -1],
                  [-3, -1, 2],
                  [-2, 1, 2]])

[u,s,v]=linearalgebra.svd(arr)

print("Orthogoanl U: \n",u)
print()
print("Diagonal sigma: \n",s)
print()
print("Orthogonal v: \n",v)
```

```python
arr
```

```python
print("Diagonals: ",np.diagonal(arr))
print("Traces: ",np.trace(arr))
```
